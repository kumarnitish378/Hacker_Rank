```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
from random import randint
from sklearn.model_selection import train_test_split
```

```python
x = np.linspace(0, np.pi, 100)
y = np.sin(4*x)
plt.plot(x, y)
```

```python
def get_sig():
  num = randint(1,10)
  return np.sin(num*x), num
```

```python
plt.plot(x, get_sig()[0])
```

```python
df = []
rs = []
for _ in range(10000):
  rt = get_sig()
  df.append(rt[0])
  rs.append(rt[1])

df = np.array(df)
rs = np.array(rs)

x_train = df[:9000]
x_test = df[9000:]
y_train = rs[:9000]
y_test = rs[9000:]
```

```python
ig = np.array([[randint(90,100) for _ in range(100)] for _ in range(10000)])/100
res = np.array([randint(0,10) for _ in range(10000)])
```

```python
x_train = ig[:9000]
x_test = ig[9000:]
y_train = res[:9000]
y_test = res[9000:]
```

```python
x_train[0], y_train[0]
```

```python
plt.plot(x_train[0])
```

```python
len(set(y_train))
```

```python
model = tf.keras.models.Sequential([
                                   tf.keras.layers.Flatten(input_shape=x_train[0].shape),
                                   tf.keras.layers.Dense(120, activation="sigmoid"),
                                   tf.keras.layers.Dense(120, activation="softmax"),
                                   tf.keras.layers.Dropout(0.2),
                                   tf.keras.layers.Dense(10, activation="tanh")
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
x_train[0].shape
```

```python
model.fit(x_train, y_train, epochs=20)
```

```python
def get_add():
  num1 = randint(1,10)
  num2 = randint(1,10)
  return [num1, num2], num1 + num2
```

```python
number = []
sum = []
for _ in range(10000):
  rt = get_add()
  number.append(rt[0])
  sum.append(rt[1])

number = np.array(number)
sum = np.array(sum)

x_train = number[:9000]
x_test = number[9000:]
y_train = sum[:9000]
y_test = sum[9000:]
```

```python
ad_model = tf.keras.models.Sequential([
                                   tf.keras.layers.Flatten(input_shape=x_train[0].shape),
                                   tf.keras.layers.Dense(120, activation="sigmoid"),
                                   tf.keras.layers.Dropout(0.2),
                                   tf.keras.layers.Dense(19, activation="tanh")
])
```

```python
ad_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
ad_model.summary()
```

```python
ad_model.fit(x_train, y_train, epochs=20)
```

```python
ad_model.evaluate(x_test, y_test)
```
